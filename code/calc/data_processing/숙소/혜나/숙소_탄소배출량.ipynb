# 각 숙소(주소)별 탄소배출량


```python
emp = 0
exi = 0
arr = pd.Series()
for i, name in zip(acc_crawling['주소'], acc_crawling['이름']):
    if c3.loc[c3['ROAD_NM_ADDR']==i].empty:
        emp += 1
        continue
    else:
        exi += 1
        arr[name] = (c3.loc[c3['ROAD_NM_ADDR']==i, 'SUM_GRGS_DSAMT'].mean())
        print(name)
        print(c3.loc[c3['ROAD_NM_ADDR']==i, 'SUM_GRGS_DSAMT'].mean())
print(f'empty : {emp}, exist : {exi}')
```

# 각 숙박 업종별 탄소 배출량
숙박 종류 (추후 1인당 탄소 배출량으로)
- 호텔
- 리조트
- 모텔
- 펜션
- 빌라
- 캠핑
- 글램핑
- 게스트하우스

```python
co2_arr = pd.DataFrame(columns={'숙박업종','co2'})
co2_arr.loc[0,'숙박업종'] = '호텔'
co2_arr.loc[1,'숙박업종'] = '리조트'
co2_arr.loc[2,'숙박업종'] = '모텔'
co2_arr.loc[3,'숙박업종'] = '펜션'
co2_arr.loc[4,'숙박업종'] = '캠핑&글램핑'
co2_arr.loc[5,'숙박업종'] = '게스트하우스'
co2_arr.loc[0,'co2'] =arr[arr.index.str.contains('호텔')].mean().values

co2_arr['호텔']=arr[arr.index.str.contains('호텔')].mean()
co2_arr['리조트']=arr[arr.index.str.contains('리조트')].mean()
co2_arr['모텔']=arr[arr.index.str.contains('모텔')].mean()
co2_arr['펜션']=arr[arr.index.str.contains('펜션')].mean()
co2_arr['캠핑&글램핑']=arr[(arr.index.str.contains("캠핑|글램핑"))].mean()
co2_arr['게스트하우스']=arr[arr.index.str.contains('게스트하우스')].mean()
co2_arr['전체']=arr.values.mean()
```
