# 음식점

```python
# 태그 기반 음식점 카테고리 분류
def tag_split(x):
    x = x.replace('\n','')
    lst = []
    for t in x.split('#'):
        lst.append(t.strip())
    lst.remove('')
    return lst

food['tag_lst'] = food['태그'].apply(lambda x: tag_split(x))

tags = []
for i in food['tag_lst'].to_list():
    for j in i:
        tags.append(j)
tags_set = set(tags)

# 회 관련 태그
회 = []
for i in tags_set:
    if ('회' in i) or ('회집' in i) or ('홍어' in i):
        회.append(i)
회.remove('회수동')
회.remove('구삼대회관')
회.remove('회전초밥')

# 흑돼지 관련 태그
흑돼지 = []
for i in tags_set:
    if ('흑돼지' in i) or ('돼지' in i) or ('오겹살' in i):
        흑돼지.append(i)
        
# 초밥 관련 태그
초밥 = []
for i in tags_set:
    if '초밥' in i:
        초밥.append(i)

# 돈카츠 관련 태그
돈카츠 = []
for i in tags_set:
    if ('돈까스' in i) or ('돈카츠' in i) or ('돈가스' in i) or ('고로케' in i):
        돈카츠.append(i)
돈카츠.remove('흑돼지돈까스')
돈카츠.append('생선카츠')

# 고기국수 관련 태그
고기국수 = []
for i in tags_set:
    if ('고기국수' in i) or ('고기 국수' in i):
        고기국수.append(i)
        
# 생선 관련 태그
생선 = []
for i in tags_set:
    if ('생선' in i) or ('구이' in i) or ('조림' in i):
        생선.append(i)
생선 = [x for x in 생선 if x not in ['특양구이','돼지구이','대창구이','말고기구이','흑돼지꼬치구이','닭모래주머니구이',
                                    '소금구이','참숯구이','초벌구이','생구이','한우구이','찰떡구이','연탄구이',
                                    '스팸구이','흑돼지숯불구이','가래떡구이','전복버터구이','숯불구이','고기구이',
                                    '돼지고기구이']]

# 이탈리안 레스토랑 관련 태그
이탈리안 = []
for i in tags_set:
    if ('이탈리안' in i) or ('파스타' in i) or ('리조또' in i) or ('피자' in i) or ('스테이크' in i):
        이탈리안.append(i)
        
# 햄버거 관련 태그
햄버거 = []
for i in tags_set:
    if ('햄버거' in i) or ('버거' in i):
        햄버거.append(i)
        
# 한식 관련 태그
한식 = []
for i in tags_set:
    if ('해물탕' in i) or ('전골' in i) or ('불백' in i) or ('도시락' in i) or ('돌솥' in i) or ('찌개' in i) or ('쌈밥' in i) or ('파전' in i) or ('백숙' in i) or ('미역국' in i) or ('해장국' in i) or ('정식' in i) or ('갈비찜' in i) or ('냉면' in i) or ('한식' in i) or ('수육' in i):
        한식.append(i)
        
# 뚝배기 관련 태그
뚝배기 = []
for i in tags_set:
    if ('설렁탕' in i) or ('뚝배기' in i) or ('국밥' in i) or ('순대국' in i):
        뚝배기.append(i)

# 칼국수 관련 태그
칼국수 = []
for i in tags_set:
    if '칼국수' in i:
        칼국수.append(i)
        
# 오메기 관련 태그
오메기 = []
for i in tags_set:
    if '오메기' in i:
        오메기.append(i)
오메기.remove('오메기술')

# 라면 관련 태그
라면 = []
for i in tags_set:
    if '라면' in i:
        라면.append(i)
        
# 카페 관련 태그
카페 = []
for i in tags_set:
    if ('떡' in i) or ('모찌' in i) or ('밀크티' in i) or ('주스' in i) or ('카페' in i) or ('케이크' in i) or ('도넛' in i) or ('빵' in i) or ('커피' in i) or ('에이드' in i) or ('아이스크림' in i) or ('아메리카노' in i) or ('차' in i) or ('빙수' in i):
        카페.append(i)
        
# 술 관련 태그
술 = []
for i in tags_set:
    if ('펍' in i) or ('칵테일' in i) or ('술' in i) or ('주류' in i) or ('전통주' in i):
        술.append(i)
술.remove('술빚기체험')

# 기타 관련 태그
기타 = []
for i in tags_set:
    if ('분식' in i) or ('인도' in i) or ('만두' in i) or ('무슬림' in i) or ('덮밥' in i) or ('짬뽕' in i) or ('라멘' in i) or ('족발' in i) or ('마라탕' in i) or ('일식' in i) or ('닭' in i) or ('뷔페' in i) or ('떡볶이' in i) or ('짜장' in i) or ('딤섬' in i) or ('다이닝' in i) or ('쌀국수' in i) or ('김밥' in i) or ('치킨' in i) or ('멕시칸' in i) or ('샤브샤브' in i):
        기타.append(i)

# 회
food['회'] = 0
idx = food[food['태그'].str.contains('|'.join(회))].index
food.loc[idx,'회'] = 1

# 흑돼지
food['흑돼지'] = 0
food.loc[food[food['태그'].str.contains('|'.join(흑돼지))].index,'흑돼지'] = 1

# 초밥
food['초밥'] = 0
food.loc[food[food['태그'].str.contains('|'.join(초밥))].index,'초밥'] = 1

# 돈카츠
food['돈카츠'] = 0
food.loc[food[food['태그'].str.contains('|'.join(돈카츠))].index,'돈카츠'] = 1

# 고기국수
food['고기국수'] = 0
food.loc[food[food['태그'].str.contains('|'.join(고기국수))].index,'고기국수'] = 1

# 생선
food['생선'] = 0
food.loc[food[food['태그'].str.contains('|'.join(생선))].index,'생선'] = 1

# 이탈리안
food['이탈리안'] = 0
food.loc[food[food['태그'].str.contains('|'.join(이탈리안))].index,'이탈리안'] = 1

# 햄버거
food['햄버거'] = 0
food.loc[food[food['태그'].str.contains('|'.join(햄버거))].index,'햄버거'] = 1

# 한식
food['한식'] = 0
food.loc[food[food['태그'].str.contains('|'.join(한식))].index,'한식'] = 1

# 뚝배기
food['뚝배기'] = 0
food.loc[food[food['태그'].str.contains('|'.join(뚝배기))].index,'뚝배기'] = 1

# 칼국수
food['칼국수'] = 0
food.loc[food[food['태그'].str.contains('|'.join(칼국수))].index,'칼국수'] = 1

# 오메기
food['오메기'] = 0
food.loc[food[food['태그'].str.contains('|'.join(오메기))].index,'오메기'] = 1

# 라면
food['라면'] = 0
food.loc[food[food['태그'].str.contains('|'.join(라면))].index,'라면'] = 1

# 카페
food['카페'] = 0
food.loc[food[food['태그'].str.contains('|'.join(카페))].index,'카페'] = 1

# 술
food['술'] = 0
food.loc[food[food['태그'].str.contains('|'.join(술))].index,'술'] = 1

# 기타
food['기타'] = 0
food.loc[food[food['태그'].str.contains('|'.join(기타))].index,'기타'] = 1

food['cat_num'] = food['회']+food['흑돼지']+food['초밥']+food['돈카츠']+food['고기국수']+food['생선']+food['이탈리안']+ food['햄버거'] + food['한식'] + food['뚝배기'] + food['칼국수'] + food['오메기'] + food['라면'] + food['카페'] + food['술'] + food['기타']
food
```
