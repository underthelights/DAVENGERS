```python
# 태그, 주소, 좋아요, 찜하기, 리뷰, 방문했어요, 조회, SNS공유

def info1():
    
    tags = driver.find_elements(By.CLASS_NAME,'tag_area')
    태그.append(tags[0].text)

    주소.append(driver.find_element(By.CLASS_NAME,'info_sub_cont').text)


    infos = driver.find_elements(By.CLASS_NAME,'appraisal_cnt')
    좋아요.append(infos[0].text)
    찜하기.append(infos[1].text)
    리뷰.append(infos[2].text)
    방문했어요.append(infos[4].text)
    조회.append(infos[5].text)
    SNS공유.append(infos[6].text)
```

```python
def info2():

    lst = ['소개','상세 정보','이용 시간','요금 정보','장소 특성','주요목적','주요목적 기타','평균 소요 시간','경사도(난이도)','편의시설','기타 상세']
    lst1 = [소개, 상세정보, 이용시간, 요금정보, 장소특성, 주요목적, 주요목적기타, 평균소요시간, 경사도난이도, 편의시설, 기타상세 ]
    titles = driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div[2]/div[2]/div[2]/div[2]/div/dl/dt')
    contents = driver.find_elements(By.XPATH,'//*[@id="content"]/div[2]/div[2]/div[2]/div[2]/div[2]/div/dl/dd')

    for l in range(len(lst)):
        for k in range(len(titles)):
            if lst[l] == titles[k].text:
                lst1[l].append(contents[k].text) 

        if len(lst1[l])!=len(이름): # 해당정보 없는 경우 빈칸 입력
            lst1[l].append("")
```

```python
# 주변관광지, 주변음식점, 주변숙박

def near_places():

    # 하단에'지도/주변관광지' 클릭
    driver.find_element(By.XPATH, '//*[@id="tab2"]/h2/a/span').click()
    time.sleep(1)


    # 주변관광지 
    driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[2]/a').click()
    time.sleep(1)

    cnt1 = (driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[2]/a')).text
    cnt1 = int(re.sub(r'[^0-9]', '', cnt1)) # 숫자만 뽑아서 저장
    주변관광지개수.append(cnt1) 

    places1 = []
    for a in range(1,cnt1+1):
        places1.append(driver.find_element(By.XPATH, '//*[@id="stab2"]/div[2]/div/div[2]/ul/li['+str(a)+']/div/dl/dd/h3/a').text)
    주변관광지.append(places1)


    # 주변음식점 
    driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[3]/a').click()
    time.sleep(1)

    cnt2 = (driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[3]/a')).text
    cnt2 = int(re.sub(r'[^0-9]', '', cnt2)) # 숫자만 뽑아서 저장
    주변음식점개수.append(cnt2) 

    places2 = []
    for b in range(1,cnt2+1):
        places2.append(driver.find_element(By.XPATH, '//*[@id="stab2"]/div[2]/div/div[2]/ul/li['+str(b)+']/div/dl/dd/h3/a').text)
    주변음식점.append(places2)
    
    
    # 주변숙박
    driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[4]/a').click()
    time.sleep(1)

    cnt3 = (driver.find_element(By.XPATH, '//*[@id="stab2"]/div[1]/div[1]/ul/li[4]/a')).text
    cnt3 = int(re.sub(r'[^0-9]', '', cnt3)) # 숫자만 뽑아서 저장
    주변숙박개수.append(cnt3) 

    places3 = []
    for c in range(1,cnt3+1):
        places3.append(driver.find_element(By.XPATH, '//*[@id="stab2"]/div[2]/div/div[2]/ul/li['+str(c)+']/div/dl/dd/h3/a').text)
    주변숙박.append(places3)
```

```python
for i in range(1,93): # 1~79페이지
    
    url = "https://www.visitjeju.net/kr/detail/list?menuId=DOM_000001718000000000&cate1cd=cate0000000002#p"+str(i)+"&pageSize=12&sortListType=reviewcnt&viewType=list&isShowBtag&tag"
    driver.get(url) # url 로 이동
    
    for j in range(1,13): # 한 페이지에 숙소 12개
    
        try:

            이름.append(driver.find_element(By.XPATH,'//*[@id="content"]/div/div[2]/div[5]/ul/li['+str(j)+']/dl/dt/a/p[1]').text) 
            지역.append(driver.find_element(By.XPATH,'//*[@id="content"]/div/div[2]/div[5]/ul/li['+str(j)+']/dl/dt/a/p[2]').text) 
            
        except NoSuchElementException: 
            break 
            
            
        # 상세 페이지 클릭
        driver.find_element(By.XPATH,'//*[@id="content"]/div/div[2]/div[5]/ul/li['+str(j)+']/dl/dt/a').click()
        time.sleep(1)
        
        
        info1()
        info2()
        상세텍스트.append(driver.find_elements(By.ID,'tab0')[0].text)
        near_places()

        # 이전 페이지로 돌아가기

        driver.back()
        time.sleep(2)
```
