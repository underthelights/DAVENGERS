```python
import pandas as pd
from itertools import combinations
import numpy as np

recommended_tour = np.array(['A2', 'A3', 'A4', 'A5', 'A12', 'A18'])
index = np.array(['A0', 'A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15', 'A16', 'A17', 'A18', 'A19'])

data = np.random.rand(20, 20)*1000
df_dist = pd.DataFrame(data = data, index = index, columns = index)
df_dist = df_dist+df_dist.T
for i in df_dist.index:
    df_dist.loc[i, i] = 0
    
df_dist
```

```python
for comb in combinations(recommended_tour, 2):
    print(comb, df_dist.loc[comb[0], comb[1]])
```
